```python
import scanpy as sc
import os
import multiprocessing
```

```python
os.chdir('/rd2/user/xiacr/sle/data/10x_rna/')
```

```python
def h5_to_h5ad(i):
    if 'h5' in i and 'bak' not in i:
        print(i)
        adata = sc.read_10x_h5(i)
        filename = '../10x_h5ad/'+i.split('_')[0]+'.h5ad'
        adata.write(filename=filename)
    else: 
        print('skip  ',i)
```

```python
process = multiprocessing.Pool(24)
target = os.listdir('/rd2/user/xiacr/sle/data/10x_rna/')
out = process.map(h5_to_h5ad, target)
process.close()
process.join()
```
